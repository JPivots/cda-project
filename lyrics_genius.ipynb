```python
# package that performs scraping of genius website to obtain lyrics and meta data
from lyricsgenius import Genius

#config file with token from genius
import config

token = config.lyricsgenius['token']
genius = Genius(token)
# Remove section headers (e.g. [Chorus]) from lyrics when searching
genius.remove_section_headers = True
# Exclude songs with these words in their title
genius.excluded_terms = ["(Remix)", "(Live)"]

artist = genius.search_artist("Taylor Swift", 2, sort="popularity")
print(artist.songs)
print(artist)

```

```python
song = genius.search_song("mean", artist.name)
print(song.lyrics)
```

```python
song.save_lyrics()
```

```python
artist.add_song(song)
artist.save_lyrics()
```
